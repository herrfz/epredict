```python
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.ensemble import RandomForestClassifier
from sklearn import cross_validation
from sklearn.preprocessing import scale, MinMaxScaler
from scipy.signal import detrend
```

```python
# utility function #1: append zero to time
def append_zero(x):
    '''input x is a string
    '''
    if len(x)==5:
        y = '0' + x
    else:
        y = x
        
    return y
```

```python
# utility function #2: detect if some events are observed
def is_event(obj):
    '''input obj is a dataframe with 4 rows and contains a 'Close' column
    '''
    if (obj.ix[1,'Close'] > obj.ix[0,'Close']) and \
       (obj.ix[2,'Close'] > obj.ix[1,'Close']) and \
       (obj.ix[3,'Close'] > obj.ix[2,'Close']) and \
       (obj.ix[3,'Close'] - obj.ix[0,'Close']) > 3. :
        y = 1.
    else:
        y = 0.
        
    return y
```

```python
df = pd.read_csv('/Users/erriza/Dropbox/machine/ES #F_M5.CSV')
times = ['Date', 'Time']
df[times] = df[times].applymap(lambda x: str(x))
df['Time'] = df['Time'].apply(append_zero)

df['Volume'] = df['Volume'].astype(float)

# create a datetime index, drop original date and time columns
df.index = (df['Date'] + df['Time']).apply(lambda x: datetime.strptime(x, '%Y%m%d%H%M%S'))
df = df.drop(times, axis=1)
df[:10]
```

```python
# change the values of invalid data to NaN for missing value handling as below
features = ['Open', 'High', 'Low', 'Close', 'Volume']
df.ix[df['Valid']==0, features] = NaN

# fill up missing values:
# first fill forward (fill missing data with current), 
# then backward (data in first rows don't have current value, so fill from the 'future')
df = df.fillna(method='ffill').fillna(method='bfill')
df[df['Valid']==0].head()
```

```python
# add some derived features
## difference of close values
df['dclose'] = df['Close'].diff().fillna(0.)
## rolling/moving average
df['nmeanclose'] = pd.rolling_mean(df['Close'], 10).fillna(method='bfill')

# add the event definition
ev = []
for ts in xrange(len(df.index)):
    try:
        if df['Close'][ts + 1] > df['Close'][ts] and \
           df['Close'][ts + 2] > df['Close'][ts + 1] and \
           df['Close'][ts + 3] > df['Close'][ts + 2] and \
           df['Close'][ts + 3] - df['Close'][ts] > 3.:
            
            ev.append(1.)
        else:
            ev.append(0.)
    except:
        ev.append(np.NaN)
        
df['event'] = ev
df[:12]
```

```python
# create complete training/test dataset (FeatureFrame)
# this cell is the bottleneck of all computations in this notebook
# look back in time = (dt - 1) * 5 minutes
# number of features = len(features) * dt
# event indicator is stored in column 'event', look forward = (lf - 1) * 5 minutes
# number of training/test examples = numdata
dt = 7
jj = dt
lf = 4
numdata = 2000

features = ['Close', 'nmeanclose']

FeatureFrame = pd.DataFrame( [tuple(np.ravel(df.ix[jj-dt:jj,features]))], index=[df.ix[jj-1].name] )
FeatureFrame['event'] = df.ix[jj-1,'event']

for i in xrange(jj+1, numdata+dt):
    temp_df = pd.DataFrame( [tuple(np.ravel(df.ix[i-dt:i,features]))], index=[df.ix[i-1].name] )
    FeatureFrame = pd.concat( [FeatureFrame, temp_df] )
    FeatureFrame.ix[-1,'event'] = df.ix[i-1,'event']
    
FeatureFrame.shape
FeatureFrame.ix[:5,-10::]
```

```python
# check percentage of events in the data
print '%.3f%%' % (float(len(FeatureFrame[FeatureFrame['event']==1.])) / float(len(FeatureFrame)) * 100)
```

```python
# split into training and test set; TBD: use sklearn cross validation
numtrain = .7 * numdata

vols = [x for x in range(dt * len(features)) if (x%len(features))==1]
ohlc = [x for x in range(dt * len(features)) if x not in vols]

train = FeatureFrame.ix[:numtrain,:-1].copy()
#train.ix[:,ohlc] = scale(detrend(train.ix[:,ohlc]))
train.ix[:,ohlc] = MinMaxScaler(feature_range=(0,1)).fit_transform(train.ix[:,ohlc])
train.ix[:,vols] = MinMaxScaler(feature_range=(0,1)).fit_transform(train.ix[:,vols])
target = FeatureFrame.ix[:numtrain,'event']

test = FeatureFrame.ix[numtrain:,:-1].copy()
#test.ix[:,ohlc] = scale(detrend(test.ix[:,ohlc]))
test.ix[:,ohlc] = MinMaxScaler(feature_range=(0,1)).fit_transform(test.ix[:,ohlc])
test.ix[:,vols] = MinMaxScaler(feature_range=(0,1)).fit_transform(test.ix[:,vols])
test_target = FeatureFrame.ix[numtrain:,'event']
```
